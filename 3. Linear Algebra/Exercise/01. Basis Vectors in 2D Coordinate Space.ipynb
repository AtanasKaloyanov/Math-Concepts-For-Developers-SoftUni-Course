```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_vectors(vectors, colors):
    for i in range (0, len(vectors)):
        vector = vectors[i]
        x_start = vector[0]
        y_start = vector[1]
        x_point = vector[2]
        y_point = vector[3]
        
        color = colors[i]
        plt.quiver(x_start, y_start, x_point - x_start, y_point - y_start, angles = "xy", scale_units = "xy", scale = 1, color = color)
   
    plt.xticks(range(-5, 10))
    plt.yticks(range(-5, 10))
    
    ax = plt.gca()
    ax.set_aspect("equal")
    ax.axhline(y = 0, color = "k")
    ax.axvline(x = 0, color = "k")
    plt.grid()
    plt.show()
```

```python
plot_vectors([[0, 0, 2, 3]], ["red"])
```

```python
plot_vectors([[1, 1, -2, 3], [2, 1, -2.5, 1.5], [-3.2, -1.5, 0, 4.3]], ["red", "blue", "orange"])
```

```python
def find_linear_combination_coefficients(e1, e2, v):
    e1_x = e1[0]
    e1_y = e1[1]
    e2_x = e2[0]
    e2_y = e2[1]
    v_x = v[0]
    v_y = v[1]
    
    A = np.array([e1, e2]).T
    B = v
    return np.linalg.solve(A, B)
# First vector coordinates: e1_x, e1_y, 
# Second vector coordinates: e2_x, e2_y  
# Linear combination coordinates: v_x, v_y
            
```

```python
e1, e2 = [[1, 0], [0, 1]]
v = [3.5, 8.6]

coefficients = find_linear_combination_coefficients(e1, e2, v)
print("Coefficients: ", str(coefficients))

plot_vectors([[0, 0, i[0], i[1]] for i in [e1, e2, v]], ["red", "blue", "green"])
```
