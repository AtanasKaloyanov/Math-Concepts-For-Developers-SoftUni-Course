```python
import matplotlib.pyplot as plt
```

```python
def plot_complex_number(z):
    plt.quiver(0, 0, z.real, z.imag, angles = "xy", scale_units = "xy", scale = 1, color = "green")
   
    plt.xticks(range(-3, 4))
    plt.yticks(range(-3, 4))
    plt.xlabel("real")
    plt.ylabel("imaginary")
   
    ax = plt.gca()
    ax.set_aspect("equal")
    plt.grid()
    plt.show()
```

```python
z = 2 + 3j
plot_complex_number(z)
```

```python
def plot_complex_numbers(numbers, colors):

    zeros = [0 for _ in numbers]
    real_nums = [n.real for n in numbers]
    imag_nums = [n.imag for n in numbers]
    min_x, max_x = int(min(real_nums))-1, int(max(real_nums)+2)
    min_y, max_y = int(min(imag_nums))-1, int(max(imag_nums)+2)
    
    plt.quiver(zeros, zeros, real_nums, imag_nums, angles = "xy", scale_units = "xy", scale = 1, color=colors)
    plt.xticks(range(min_x, max_x))
    plt.yticks(range(min_y, max_y))
    plt.gca().set_aspect("equal")
    plt.grid()
    plt.show()
    
plot_complex_numbers([2 + 3j, -2 - 1j, -3, 2j], 
                     ["green", "red", "blue", "orange"])
    
    
```

```python
z1 = 2 + 3j
z2 = 1 - 1j
plot_complex_numbers([z1, z2, z1 + z2], ["red", "blue", "green"])
```

```python
z1 = 2 + 3j
z2 = 2 + 0j
plot_complex_numbers([z1, z2, z1 + z2], ["red", "blue", "green"])
```

```python
z1 = 2 + 3j
z2 = 0 + 2j
plot_complex_numbers([z1, z2, z1 + z2], ["red", "blue", "green"])
```

```python
z = 2 + 3j
plot_complex_numbers([z, z * 1], ["red", "blue"])
plot_complex_numbers([z, z * -1], ["red", "blue"])
plot_complex_numbers([z, z * 1j], ["red", "blue"])
plot_complex_numbers([z, z * -1j], ["red", "blue"])
```

```python
z1 = 2 + 3j
z2 = 1 - 2j
plot_complex_numbers([z1, z2, z1 * z2], ["red", "blue", "green"])
```
